### Pre-processing data

```python
# Cell 1: Import required libraries
import pandas as pd
import numpy as np
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from transformers import BertTokenizer
import torch
from sklearn.preprocessing import MinMaxScaler
from datetime import datetime
import os

# Download required NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
```

```python
# Cell 2: Define the ReviewPreprocessor class
class ReviewPreprocessor:
    def __init__(self, max_length=512):
        self.tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
        self.max_length = max_length
        self.scaler = MinMaxScaler()
        self.stop_words = set(stopwords.words('english'))
        
    def clean_text(self, text):
        """Clean and normalize text data"""
        if not isinstance(text, str):
            return ""
            
        # Convert to lowercase
        text = text.lower()
        
        # Remove URLs
        text = re.sub(r'http\S+|www\S+|https\S+', '', text)
        
        # Remove special characters and numbers but keep basic punctuation
        text = re.sub(r'[^a-zA-Z\s.,!?]', '', text)
        
        # Remove extra whitespace
        text = ' '.join(text.split())
        
        return text
        
    def normalize_numerical_features(self, df):
        """Normalize numerical features like ratings and likes"""
        df['normalized_rating'] = self.scaler.fit_transform(df[['rating']])
        df['normalized_likes'] = self.scaler.fit_transform(df[['likes']])
        return df
        
    def process_dates(self, df):
        """Process and normalize date information"""
        df['processed_date'] = pd.to_datetime(df['iso_date'])
        df['days_since_review'] = (datetime.now() - df['processed_date']).dt.days
        df['normalized_recency'] = self.scaler.fit_transform(df[['days_since_review']])
        return df
        
    def tokenize_for_bert(self, text):
        """Tokenize text for BERT model"""
        encoded = self.tokenizer.encode_plus(
            text,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoded['input_ids'],
            'attention_mask': encoded['attention_mask']
        }
        
    def create_feature_vector(self, row):
        """Create a combined feature vector for each review"""
        features = []
        features.extend([
            row['normalized_rating'],
            row['normalized_likes'],
            row['normalized_recency']
        ])
        return np.array(features)
    
    def preprocess_dataset(self, df):
        """Complete preprocessing pipeline"""
        # Create a copy to avoid modifying the original
        processed_df = df.copy()
        
        # Clean text fields
        print("Cleaning text fields...")
        processed_df['cleaned_title'] = processed_df['title'].apply(self.clean_text)
        processed_df['cleaned_snippet'] = processed_df['snippet'].apply(self.clean_text)
        processed_df['cleaned_response'] = processed_df['response'].apply(self.clean_text)
        
        # Normalize numerical features
        print("Normalizing numerical features...")
        processed_df = self.normalize_numerical_features(processed_df)
        
        # Process dates
        print("Processing dates...")
        processed_df = self.process_dates(processed_df)
        
        # Create BERT tokens for text fields
        print("Creating BERT tokens...")
        processed_df['title_tokens'] = processed_df['cleaned_title'].apply(self.tokenize_for_bert)
        processed_df['snippet_tokens'] = processed_df['cleaned_snippet'].apply(self.tokenize_for_bert)
        
        # Create feature vectors
        print("Creating feature vectors...")
        processed_df['feature_vector'] = processed_df.apply(self.create_feature_vector, axis=1)
        
        return processed_df
    
    def prepare_bert_inputs(self, processed_df, text_column='cleaned_snippet'):
        """Prepare inputs for BERT model"""
        return self.tokenizer.batch_encode_plus(
            processed_df[text_column].tolist(),
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
```
