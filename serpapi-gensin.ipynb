```python
import pandas as pd
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Download stopwords dan punkt jika belum di-download
nltk.data.path.append('/home/rebecca/nltk_data')
nltk.download('punkt', download_dir='/home/rebecca/nltk_data')
nltk.download('stopwords', download_dir='/home/rebecca/nltk_data')

# Membaca dataset
file_path = 'data/google-play-rev-gen-2.csv'
df = pd.read_csv(file_path)
```

```python
# Inisialisasi objek stopwords dan stemmer
stop_words = set(stopwords.words('indonesian'))  # untuk Bahasa Indonesia
stemmer = PorterStemmer()  # Bisa ganti dengan Sastrawi untuk Bahasa Indonesia

# Melakukan pemrosesan langsung pada kolom 'snippet'
df['processed_snippet'] = df['snippet'].apply(lambda text: 
    # 1. Menghapus karakter khusus, angka, dan tanda baca
    re.sub(r'[^a-zA-Z\s]', '', text)  # Menghapus angka dan karakter non-alfabet
    .lower()  # Mengubah semua huruf menjadi huruf kecil
    # 2. Tokenisasi
    .apply(word_tokenize)  # Tokenisasi
    # 3. Menghapus stopwords
    .apply(lambda tokens: [word for word in tokens if word not in stop_words])
    # 4. Stemming
    .apply(lambda tokens: [stemmer.stem(word) for word in tokens])
    # Menggabungkan kembali token yang sudah diproses menjadi teks
    .apply(lambda tokens: ' '.join(tokens))
)

# Menampilkan beberapa hasil pemrosesan
print(df[['id', 'snippet', 'processed_snippet']].head())

# Menyimpan hasil ke file baru (opsional)
df.to_csv('data/google-play-rev-gen-2-processed.csv', index=False)
```
